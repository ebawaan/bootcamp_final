```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# How has BMI of winners changed over the years
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn scatterplot
sns.scatterplot(data=Data1, x='Year', y='BMI')

plt.title('BMI vs Year of Winner')
plt.xlabel('Year')
plt.ylabel('BMI')
plt.grid(True)
plt.show()
```

This plot shows a general downward trend in BMI of the winners of the tour as time has progressed

```python
plt.scatter(Data1['Year'], Data1['weight_(Kg)'], marker='o')
plt.xlabel('Year')
plt.ylabel('weight_kg')
plt.title('weight vs Year')
plt.grid(True)
plt.show()
```

```python
# What role does height play?
plt.scatter(Data1['Year'], Data1['height_(m)'], marker='o')
plt.xlabel('Year')
plt.ylabel('Height')
plt.title('Height vs Year')
plt.grid(True)
plt.show()
```

```python
# How has the age of the winners changed over time
plt.scatter(Data1['Year'], Data1['age'])
plt.xlabel('Year')
plt.ylabel('Age')
plt.title('Age vs Year')
plt.grid(True)
plt.show()
```

```python
# what is the # of non-stage races participated in by winners of the tour
sns.histplot(Data2['total_completed_day_races'], bins=10, edgecolor='black')

plt.xlabel('Total Completed Day Races')
plt.ylabel('Frequency')
plt.title('Histogram of Total Completed Day Races')
plt.show()
```

```python
# How has the participation in non-stage race been throughout the years
plt.scatter(Data2['Year'], Data2['total_completed_day_races'])
plt.xlabel('Year')
plt.ylabel('Total Completed Day Races')
plt.title('Total Completed Day Races vs Year')
plt.grid(True)
plt.show()
```

```python
# What is the ideal rider type?

# Clean the rider_type_(PPS) column by stripping whitespace
Data1['rider_type_(PPS)_clean'] = Data1['rider_type_(PPS)'].str.strip()

# Now plot
sns.countplot(data=Data1, x='rider_type_(PPS)_clean', palette='Set2', edgecolor='black')

plt.xlabel('Rider Type (PPS)')
plt.ylabel('Count')
plt.title('Count of Each Rider Type (Cleaned)')
plt.show()
```

```python
# How has the length of the tour changed over the years
Data1['Tour_overall_length_(km)'] = pd.to_numeric(Data1['Tour_overall_length_(km)'], errors='coerce')

sns.scatterplot(data=Data1, x='Year', y='Tour_overall_length_(km)')

plt.xlabel('Year')
plt.ylabel('Tour Length (km)')
plt.title('Tour Length vs Year')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=Data1, x='BMI', y='Tour_overall_length_(km)')
plt.title('Scatterplot of BMI vs. Tour Length')
plt.xlabel('BMI')
plt.ylabel('Tour Length (km)')
plt.show()
```
